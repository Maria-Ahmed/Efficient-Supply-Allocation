```python
# Finding How many orders are coming in specific areas(lat,longitude)
start_unique_area_order_count= data.groupby(['start_lat','start_lng']).size().reset_index().rename(columns={0:'count'})
```

```python
# finding how much value is generated by specific area in terms of price.
start_unique_area_order_price= data.groupby(['start_lat','start_lng'])['ride_value'].sum().reset_index().rename(columns={0:'count'})
```

```python
#finding unique areas and the timestamp of the order
start_unique_timestamps= data.groupby(['start_lat','start_lng'])['Time'].unique().reset_index().rename(columns={0:'count'})
```

```python
# Combining (areas by order count) and (areas by order price)

df_count_price = pd.merge(start_unique_area_order_count, start_unique_area_order_price,  how='inner', 
                          left_on=['start_lat','start_lng'], right_on = ['start_lat','start_lng'])
```

```python
# merge df_count_price with "start_unique_timestamps" to create a base table(df_count_price_time)

df_count_price_time = pd.merge(df_count_price, start_unique_timestamps,  how='inner', 
                               left_on=['start_lat','start_lng'], right_on = ['start_lat','start_lng'])
```

```python
df_count_price_time
```

```python
# sorted the base table W.R.T count & ride_value : giving priority to ride counts

df_count_price_time_sort= df_count_price_time.sort_values(['ride_value', 'count'], ascending=False).head(10)
```

```python
X = data[["start_lat","start_lng"]]
max_k = 10

## iterations
distortions = [] 
for i in range(1, max_k+1):
    if len(X) >= i:
       model = cluster.KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
       model.fit(X)
       distortions.append(model.inertia_)
    
## best k: the lowest derivative
k = [i*100 for i in np.diff(distortions,2)].index(min([i*100 for i 
     in np.diff(distortions,2)]))

## plot
fig, ax = plt.subplots()
ax.plot(range(1, len(distortions)+1), distortions)
ax.axvline(k, ls='--', color="red", label="k = "+str(k))
ax.set(title='The Elbow Method', xlabel='Number of clusters', 
       ylabel="Distortion")
ax.legend()
ax.grid(True)
plt.show()
```

```python
k = 6
model = cluster.KMeans(n_clusters=k, init='k-means++')
X = data[["start_lat","start_lng"]]
```

```python
## clustering 
data_X = X.copy()
data_X["cluster"] = model.fit_predict(X)
```

```python
## clustering
data_X = X.copy()
data_X["cluster"] = model.fit_predict(X)

## find real centroids
closest, distances = scipy.cluster.vq.vq(model.cluster_centers_, 
                     data_X.drop("cluster", axis=1).values)
data_X["centroids"] = 0
for i in closest:
    data_X["centroids"].iloc[i] = 1
    
## add clustering info to the original dataset
data[["cluster","centroids"]] = data_X[["cluster","centroids"]]
data.sample(5)
```
